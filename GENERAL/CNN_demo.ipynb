```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def show_img(image, title):
    
    plt.figure(figsize = (10,5))
    plt.title(title, size = 20)
    plt.imshow(image)
    plt.axis(False)
    plt.show()
```

```python
img = cv2.imread('/Users/gursi/Desktop/ML/sample_sets/random_sample1.jpg')

temp1 = img[:,:,0].copy()
temp2 = img[:,:,2].copy()

img[:,:,0] = temp2
img[:,:,2] = temp1

print(img.shape)
show_img(img, 'Original image')
```

```python
img_edges = cv2.Canny(img, 150, 150)
show_img(img_edges, 'Edge detection')
```

```python
def dilate_img(image, kernel):
    
    print('Kernel : ')
    print(kernel)
    
    fig, ax = plt.subplots(nrows = 3, ncols = 3, figsize = (17,10))
    fig.suptitle('Dilated', color = 'white', size = 20)
    fig.set_facecolor('black')

    counter = 1
    for var1 in ax :
        for var2 in var1 :    
            img_d2 = cv2.dilate(image, kernel, iterations = counter)
            var2.imshow(img_d2)
            var2.set_title(f'{counter} iterations', color = 'white')
            var2.axis(False)
            counter += 1
```

```python
k1 = np.ones(shape = (3,3))
dilate_img(img, k1)
```

```python
k2 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,3))
dilate_img(img, k2)
```

```python
k3 = np.ones(shape = (2,2))
dilate_img(img_edges, k3)
```
