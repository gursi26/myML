# Making changes to data

```python
df['Total'] = df["HP"] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
# Creates a new column with a total of all the stats
df.head(10)
```

```python
df = df.drop(['Total'], axis = 'columns')
# Drops a value or a list of values in specified axis 
```

```python
df.head()
```

```python
df['Total'] = df.iloc[:, 4:10].sum(axis = 1)
# Axis = 1 means horizontally, axis = 0 means vertically
df.head()
```

```python
df = df[['Name', 'Type 1', 'Type 2', 'Total', 'HP', 'Attack', 'Defense', 'Sp. Atk',
       'Sp. Def', 'Speed', 'Generation', 'Legendary']]
# Reordering the dataframe columns
df.head()
```
