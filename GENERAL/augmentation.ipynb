```python
from keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = img_to_array(load_img('/Users/gursi/Desktop/comp_ee/pil_test.jpg', target_size = (256,256)))/255.0
plt.imshow(img)
plt.title('Initial image', size = 25)
plt.axis(False)
plt.show()
```

```python
img = img.reshape((1,256,256,3))
data_gen = ImageDataGenerator(horizontal_flip = True, vertical_flip = True, rotation_range = 90)
data = data_gen.flow(img, batch_size = 1)
```

```python
fig, ax = plt.subplots(nrows = 3, ncols = 5, figsize = (20,10))
fig.suptitle("Augmented data", size = 30)

for var1 in ax :
    for var2 in var1 : 
        var2.imshow(next(data)[0])
        var2.axis(False)
```
