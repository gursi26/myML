```python
import pandas as pd
import matplotlib.pyplot as plt
import torch
from torch.utils.data import Dataset
import torchvision.transforms as transforms
import os,sys
from PIL import Image
import numpy as np
```

```python
current_path = os.path.dirname(os.path.abspath(sys.argv[1]))
imgs_path = current_path + '/img'
labels_path = current_path + '/index.csv'

labels = pd.read_csv(labels_path)
labels.head()
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 4, figsize = (20,8))
counter = 0

for var1 in ax :
    for var2 in var1 : 
        image_path = f'{current_path}/' + labels['image'][counter]
        img = Image.open(image_path)
        img_label = labels['category'][counter]
        
        var2.imshow(np.array(img), cmap = 'gray')
        var2.set_title(img_label)
        counter += 1
```

```python
sample_image = np.array(Image.open(current_path + '/' + labels['image'][0]))
print(sample_image.shape)
```

```python
image_tensor = torch.from_numpy(sample_image)/255.0
plt.imshow(image_tensor, cmap = 'gray')
```
