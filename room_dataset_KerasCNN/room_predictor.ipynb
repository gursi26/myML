```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
import os
from skimage import io
```

```python
model_path = "/users/gursi/desktop/ml/saved_models/clean_messy_room_acc96.h5"
pretrained_model = load_model(model_path)
```

```python
folder = '/users/gursi/desktop/ml/sample_sets/room_sample'
target_size = 256
items = os.listdir(folder)
images_arr = []

for var1 in items :
    img = image.load_img(folder + '/'+ var1, target_size = (target_size,target_size)) 
    img_arr = image.img_to_array(img).copy()
    img_arr = img_arr/255.0
    images_arr.append(img_arr)
```

```python
model_input = []
for var3 in images_arr :
    model_input.append(var3.reshape(1,target_size,target_size,3).copy())
```

```python
preds = []
classes = ['clean','messy']

for var4 in model_input :
    prediction = pretrained_model.predict(var4)
    pred_word = classes[prediction.argmax()]
    preds.append(pred_word)

print(preds)
```

```python
nrows = 0
ncols = 4
while True:
    if nrows*ncols > len(preds):
        break
    nrows = nrows + 1
```

```python
fig, ax = plt.subplots(nrows = nrows, ncols = ncols, figsize = (20,nrows*5))

pic_counter = 0
for var5 in ax :
    for var6 in var5 :
        
        try :
            var6.imshow(images_arr[pic_counter])
            var6.set_title(preds[pic_counter], fontsize = 15)
        except IndexError :
            pass
        
        var6.axis(False)
        pic_counter = pic_counter + 1
```
